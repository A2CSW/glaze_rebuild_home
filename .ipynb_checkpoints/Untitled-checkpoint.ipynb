```python
import pandas as pd

students = pd.DataFrame(columns=["student_id","name","parent_id","status"])
payments = pd.DataFrame(columns=["payment_id","student_id","amount","date","status"])
register = pd.DataFrame(columns=["student_id","class_id","date","present"])

print("DATA LAYERS INITIALISED")
```

```python
students.loc[0] = ["S001", "Test Student", "P001", "active"]
payments.loc[0] = ["PAY001", "S001", 20, "2026-06-12", "paid"]
register.loc[0] = ["S001", "C001", "2026-06-12", True]

print("SEED DATA LOADED")
```

```python
def get_payment_status(student_id):
    student_payments = payments[payments["student_id"] == student_id]
    
    if student_payments.empty:
        return "NO PAYMENTS"
    
    total_paid = student_payments["amount"].sum()
    
    if total_paid >= 20:
        return "PAID"
    elif total_paid > 0:
        return "PARTIAL"
    else:
        return "OVERDUE"

print(get_payment_status("S001"))
```

```python
def view_register():
    merged = register.merge(students, on="student_id", how="left")
    return merged

view_register()
```

```python
def attendance_summary(student_id):
    records = register[register["student_id"] == student_id]
    total = len(records)
    present = records["present"].sum()
    
    if total == 0:
        return "NO RECORDS"
    
    return {
        "total_sessions": total,
        "present": int(present),
        "attendance_rate": present / total
    }

attendance_summary("S001")
```

```python
import numpy as np

def attendance_summary_clean(student_id):
    records = register[register["student_id"] == student_id]
    total = len(records)
    present = records["present"].sum()
    
    if total == 0:
        return "NO RECORDS"
    
    return {
        "total_sessions": total,
        "present": int(present),
        "attendance_rate": round(float(present / total), 2)
    }

attendance_summary_clean("S001")
```

```python
def student_dashboard(student_id):
    s = students[students["student_id"] == student_id].iloc[0]
    pay = get_payment_status(student_id)
    att = attendance_summary_clean(student_id)
    
    return {
        "student": s["name"],
        "payment_status": pay,
        "attendance": att
    }

student_dashboard("S001")
```

```python
students.loc[1] = ["S002", "Second Student", "P002", "active"]

payments.loc[1] = ["PAY002", "S002", 10, "2026-06-12", "partial"]

register.loc[1] = ["S002", "C001", "2026-06-12", False]

print("SECOND STUDENT ADDED")
```

```python
def all_students_dashboard():
    result = []
    
    for sid in students["student_id"]:
        result.append(student_dashboard(sid))
    
    return result

all_students_dashboard()
```

```python
def risk_flags():
    output = []
    
    for sid in students["student_id"]:
        pay = get_payment_status(sid)
        att = attendance_summary_clean(sid)
        
        flag = []
        
        if pay != "PAID":
            flag.append("PAYMENT_ISSUE")
        
        if att != "NO RECORDS" and att["attendance_rate"] < 0.8:
            flag.append("ATTENDANCE_RISK")
        
        if len(flag) == 0:
            flag.append("OK")
        
        output.append({
            "student_id": sid,
            "flags": flag
        })
    
    return output

risk_flags()
```

```python
def admin_summary():
    total_students = len(students)
    
    paid = sum(get_payment_status(sid) == "PAID" for sid in students["student_id"])
    
    attendance_rates = []
    for sid in students["student_id"]:
        att = attendance_summary_clean(sid)
        if att != "NO RECORDS":
            attendance_rates.append(att["attendance_rate"])
    
    avg_attendance = sum(attendance_rates) / len(attendance_rates) if attendance_rates else 0
    
    return {
        "total_students": total_students,
        "fully_paid": paid,
        "average_attendance": round(avg_attendance, 2)
    }

admin_summary()
```

```python
students.to_csv("students.csv", index=False)
payments.to_csv("payments.csv", index=False)
register.to_csv("register.csv", index=False)

print("EXPORTED")
```
